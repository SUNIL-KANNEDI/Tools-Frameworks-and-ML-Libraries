```python
import numpy as np
import matplotlib.pyplot as plt

# Define activation functions
def sigmoid(Z):
    return 1 / (1 + np.exp(-Z))

def tanh(Z):
    return np.tanh(Z)

def relu(Z):
    return np.maximum(0, Z)

def leaky_relu(Z, alpha=0.01):
    return np.where(Z > 0, Z, alpha * Z)

def elu(Z, alpha=1.0):
    return np.where(Z > 0, Z, alpha * (np.exp(Z) - 1))

def softmax(Z):
    exp_Z = np.exp(Z - np.max(Z))  # For numerical stability
    return exp_Z / np.sum(exp_Z)

# Generate input values
# Z = np.linspace(-5, 5, 100)

Z = np.array([i for i in range(-10,10)])
# Compute function outputs
sigmoid_values = sigmoid(Z)
tanh_values = tanh(Z)
relu_values = relu(Z)
leaky_relu_values = leaky_relu(Z)
elu_values = elu(Z)

# Plot activation functions
plt.figure(figsize=(12, 8))

plt.subplot(2, 3, 1)
plt.plot(Z, sigmoid_values, label="Sigmoid", color='b')
plt.title("Sigmoid Activation")
plt.grid()

plt.subplot(2, 3, 2)
plt.plot(Z, tanh_values, label="Tanh", color='r')
plt.title("Tanh Activation")
plt.grid()

plt.subplot(2, 3, 3)
plt.plot(Z, relu_values, label="ReLU", color='g')
plt.title("ReLU Activation")
plt.grid()

plt.subplot(2, 3, 4)
plt.plot(Z, leaky_relu_values, label="Leaky ReLU", color='m')
plt.title("Leaky ReLU Activation")
plt.grid()

plt.subplot(2, 3, 5)
plt.plot(Z, elu_values, label="ELU", color='c')
plt.title("ELU Activation")
plt.grid()

plt.tight_layout()
plt.show()
```
